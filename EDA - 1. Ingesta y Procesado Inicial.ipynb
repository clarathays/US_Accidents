**Eliminado de datos**

Eliminamos las columnas consideramos que no aportan valor para el análisis 

1. End Lat
2. End Lng
3. Description
4. Number
5. Zipcode
6. Country (Son todas en el USA)
7. Timezone
8. Airport Code
9. Weather Timestap
10. Wind Direction

12. Astronomical Twilight



```python
accidentes = accidentes.drop(["End_Lat" , "End_Lng", "Description", "Number", "Zipcode" ],axis = 1)
```

```python
accidentes = accidentes.drop(["Country","Timezone", "Airport_Code", "Weather_Timestamp", "Wind_Direction","Nautical_Twilight" ,"Astronomical_Twilight"],axis = 1)
```
